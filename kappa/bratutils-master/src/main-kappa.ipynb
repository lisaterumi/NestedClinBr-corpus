```python
# fazer para todos docs e depois gerar media?
import os
# assign directory
path1=r'..\..\anotadora1'
path2=r'..\..\anotadora2'

novoFile1='all1.ann'
novoFile2='all2.ann'
 
f1 = open(novoFile1, "a", encoding='utf-8')

for filename in os.listdir(path1):
    f = os.path.join(path1, filename)
    # checking if it is a file
    if os.path.isfile(f):
        extension = os.path.splitext(filename)[1][1:]
        if extension=='ann':
            #print(f)
            file = open(f, "r", encoding='utf-8')
            linha=file.readlines()
            f1.write(''.join(linha))
f1.close()

f2 = open(novoFile2, "a", encoding='utf-8')

for filename in os.listdir(path2):
    f = os.path.join(path2, filename)
    # checking if it is a file
    if os.path.isfile(f):
        extension = os.path.splitext(filename)[1][1:]
        if extension=='ann':
            #print(f)
            file = open(f, "r", encoding='utf-8')
            linha=file.readlines()
            f2.write(''.join(linha))
f2.close()
```

```python
from bratutils import agreement as a
from importlib import reload  
a = reload(a)

doc = a.Document(novoFile1)
doc2 = a.Document(novoFile2)

#doc.make_gold() -> muda pra MISSING.. 
doc.reverse_gold() # ->  muda para SPURIOUS

# default: comp_status = MucTable.SPURIOUS e comparison_type = MucTable.STRICT_COMPARISON
statistics = doc2.compare_to_gold(doc)

print(statistics)
```

RELAXED_COMPARISON = 1

STRICT_COMPARISON = 2

BORDER_COMPARISON = 3


CORRECT - when annotation tags and indices match completely

INCORRECT - when annotation tags do not match, but the indices coincide

PARTIAL - when the annotation tags are the same but one of the annotations has the same end index and a different start index

MISSING - annotations exising only in the gold standard annotation set

SPURIOUS - annotations existing only in the candidate annotation set
